Gymnasium library: [here](https://github.com/Farama-Foundation/Gymnasium)

* An interface that allows you to create RL environments.
* A collection of environments(gym-control, atari ,..)

```python
import gymnasium as gym

# create our environment called LunarLander-v2
env = gym.make("LunarLander-v2")

# reset this environment to its initial state
observation, info = env.reset()

for _ in range(20):
    # Take a random action
    action = env.action_space.sample()
    print("Action taken:", action)

    # Do this action in the environment and get
    # next_state, reward, terminated, truncated and info
    observation, reward, terminated, truncated, info = env.step(action)

    # If the game is terminated (in our case we land, crashed) or truncated (timeout)
    if terminated or truncated:
        # Reset the environment
        print("Environment is reset")
        observation, info = env.reset()

env.close()
```

```python
# create our environment with gym.make("<name_of_the_environment>")
env = gym.make("LunarLander-v2")
env.reset()
print("_____OBSERVATION SPACE_____ \n")
print("Observation Space Shape", env.observation_space.shape)
print("Sample observation", env.observation_space.sample())  # Get a random observation
print("\n _____ACTION SPACE_____ \n")
print("Action Space Shape", env.action_space.n)
print("Action Space Sample", env.action_space.sample())  # Take a random action
```

Observation_space.shape has a vector size of 8:
* Horizontal pad coordinate (x)
* Vertical pad coordinate (y)
* Horizontal speed (x)
* Vertical speed (y)
* Angle
* Angular speed
* If left leg contact point touched the land (boolean)
* If right leg contact point touched the land (boolean)

Action space is discrete with 4:
* Action 0: Do nothing
* Action 1: Fire left orientation engine
* Action 2: Fire the main engine
* Action 3: Fire right orientation engine

Reward fn:
After every step a reward is granted. The total reward of an episode is the sum of the rewards for all the steps within that episode.

For each step, the reward:
* Is increased/decreased the closer/further the lander is to the landing pad.
* Is increased/decreased the slower/faster the lander is moving.
* Is decreased the more the lander is tilted (angle not horizontal).
* Is increased by 10 points for each leg that is in contact with the ground.
* Is decreased by 0.03 points each frame a side engine is firing.
* Is decreased by 0.3 points each frame the main engine is firing.

The episode receive an additional reward of -100 or +100 points for crashing or landing safely respectively.

An episode is considered a solution if it scores at least 200 points

Vectorized Environment
* To have diverse experiences during training, 16 envs

```python
# Create env
env = make_vec_env("LunarLander-v2", n_envs=16)
```

lib info: [here](https://stable-baselines3.readthedocs.io/en/master/)

SB3 PPO one of the state of the art DRL algorithms: [here](https://stable-baselines3.readthedocs.io/en/master/modules/ppo.html#example%5D)

PPO is a combination of:
* Value-based RL method
* Polcy-based RL method

MultiLayerPerceptron is used since the input is a vector.
If we had frames as input, use CnnPolicy

```python
# Create env (Stable-Baselines3)
env = gym.make('LunarLander-v2', render_mode='rgb_array')

# Instantiate PPO
model = PPO('MlpPolicy', env, verbose=1)

# Train agent
model.learn(total_timesteps=int(2e5))
```

* n_stpes: number of stpes to run for each env per update (rollout buffer size = n_steps*n_envs)
* batch_size: minibatch size
* n_epochs: number of epoch optimising surrogate loss
* gamma: discount factor
* gae_lambda: factor for trade-off of bias vs variance for 'GAE'
* ent_coef: entropy coef for loss calculation
* verbose: Verbosity 0 for no output (boolean)

```python
# Train the model for 100 times?
model.learn(total_timesteps=100)

# Save it
model_name = "ppo-LunarLander-v2"
model.save(model_name)
```

Evaluation
* wrap the env in a [Monitor](https://stable-baselines3.readthedocs.io/en/master/common/monitor.html)
* Check the performance with: evalute_policy
[documentation](https://stable-baselines3.readthedocs.io/en/master/guide/examples.html#basic-usage-training-saving-loading)

```python
eval_env = Monitor(gym.make("LunarLander-v2", render_mode="rgb_array"))
# Evaluate 5 epsodes and deterministic
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=5, deterministic=True)
print(f"mean of reward: {mean_reward:.4f} +/- std of reward {std_reward:.4f}")
```
